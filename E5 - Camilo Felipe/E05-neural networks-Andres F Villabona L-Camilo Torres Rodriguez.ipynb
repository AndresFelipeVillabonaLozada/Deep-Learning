#  Neural network to solve the problem In Keras

````python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
numpy.random.seed(777)
# load "The Little Red Riding Hood Neural" dataset
# split into input (X) and output (Y) variables
dataset = numpy.loadtxt("LRRHNN.csv", delimiter=",")
#X = [[1,1,0,0],[0,1,1,0],[0,0,0,1]]
#Y = [[1,0,0,0],[0,0,1,1],[0,1,1,0]]
X = dataset[:,0:4]
Y = dataset[:,4:8]
````

````python
#X
````

````python
#Y
````

````python
# Create model
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(units=4, input_dim=4))
model.add(Activation('relu'))
model.add(Dense(units=4))
model.add(Activation('sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=1100, batch_size=1)
````

````python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
````

* Train the following models:
     1. Train a linear regression model using the scaled data.
         * Report the mean square error on the test set
     2. Train a regression model using a 2-layer MultiLayer Perceptron (128 neurons in the first and 512 in the second) and with the **scaled data**.

````python
#Generate a 70-30 partitioning of the data using all the features. (Do not include the dependent variable MEDV)
#excluyendo la variable dependiente MEDV
XX = bos_data.iloc[:, :-1].values  
yy = bos_data.iloc[:, 1].values 
#partición train/test split
from sklearn.model_selection import train_test_split  
Xtrain, Xtest, ytrain, ytest = train_test_split(XX, yy, test_size=0.3, random_state=1) 
````

````python
#Train a linear regression model with the objective of predicting the output variable MEDV.
from sklearn.linear_model import LinearRegression  
lr = LinearRegression()  
lr.fit(Xtrain, ytrain) 
````

````python
#Report the mean square error on the test set
yhat = lr.predict(Xtest)  
from sklearn import metrics  
print('MSE', metrics.mean_squared_error(ytest, yhat))  
````

````python
#Train a regression model using MLPRegressor in order to predict the output variableMEDV.
#creación del modelo 
from sklearn.neural_network import MLPRegressor
lr2 = MLPRegressor(hidden_layer_sizes=(5,),
                                       activation='relu',
                                       solver='adam',
                                       learning_rate='adaptive',
                                       max_iter=100,
                                       learning_rate_init=0.01,
                                       alpha=0.01)
````

````python
#entrenamiento del modelo
lr2.fit(Xtrain, ytrain)
````

````python
#Report the mean square error on the test set
from sklearn import metrics 
yhat2 = lr2.predict(Xtest)   
print('MSE2', metrics.mean_squared_error(ytest, yhat2))  
````

````python
#Scale the data so that they have zero mean variance one per feature (only  XX ). You can use the following piece of code:
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x.fit(XX)
Xtrain_s = sc_x.transform(Xtrain)
Xtest_s = sc_x.transform(Xtest)
````

````python
#Train the following models:
#1. Tain a linear regression model using the scaled data.
from sklearn.linear_model import LinearRegression  
lr3 = LinearRegression()  
lr3.fit(Xtrain_s , ytrain) 
````

````python
#Report the mean square error on the test set
yhat3 = lr3.predict(Xtest_s)  
from sklearn import metrics  
print('MSE3', metrics.mean_squared_error(ytest, yhat3))  
````

````python
#2. Train a regression model using a 2-layer MultiLayer Perceptron (128 neurons in the first and 512 in the second) 
#and with the **scaled data**.
#libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
#model definition
def NN():
# create model
    model = Sequential()
    model.add(Dense(128, input_dim=13, kernel_initializer='normal', activation='relu'))
    model.add(Dense(512, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
````

````python
seed = 99
np.random.seed(seed)
lr4= KerasRegressor(build_fn=NN, epochs=100, batch_size=5, verbose=0)
````

````python
#Report the mean square error on the test set
#Usando valiadción cruzada
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(lr4, Xtest_s, ytest, cv=kfold)
print("Mean Squared Error LRNN: %.2f (%.2f)" % (results.mean(), results.std()))
````

**Which model has better performance? Why?** 

Con base en los resultados anteriores, se concluye que los modelos traidicionales de regresión lineal (Lr1) y (Lr3) se obtuvo el mejor MSE, entre esotos dos lr3 (con los datos escalados) presento el mejor desempeño, por lo que consideramos que este es el mejor abordaje a un problema de regresión multivariado donde las variables son lineales (ver gráficas) las NN se desempeñan mejor en modelos mas sofisticados.
